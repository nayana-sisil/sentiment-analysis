# Sentiment Analysis for Campus LMS Support System: Priority Classification

## Business Understanding

The campus Learning Management System (LMS) serves as a critical hub for student academic life, housing course materials, assignments, grades, and communication channels. Currently, student support requests ranging from technical issues like "WiFi is not working" to serious personal concerns like financial hardship are handled without an automated system to assess urgency. This lack of prioritization leads to delayed responses to critical issues, student frustration, and potential academic setbacks.

The primary business goal is to develop an intelligent sentiment analysis system that can automatically classify incoming student support reports into **High**, **Medium**, or **Low** priority categories. By implementing this system, the university can ensure that urgent issues receive immediate attention, resources are allocated efficiently, and student satisfaction and retention are improved.

### Key Benefits

| Benefit | Description |
|---------|-------------|
| **Reduced Response Time for Critical Issues** | Urgent problems like financial crises or academic dishonesty are flagged and addressed first. |
| **Optimized Resource Allocation** | Support staff can focus their efforts on high-impact issues rather than sifting through all reports manually. |
| **Improved Student Experience** | Students feel heard and supported, knowing their most pressing concerns are prioritized. |
| **Data-Driven Insights** | The university can identify recurring issues across different departments (e.g., IT, Financial Aid, Academic Advising) and address root causes proactively. |

---

## Analytic Approach

The analytic approach involves building a multi class classification model that predicts the priority level of a student support report based on its textual content. The process follows a structured machine learning pipeline:

### 1. Data Exploration and Preprocessing

- Analyze the dataset's structure, including features like:
  - Genre
  - Age
  - GPA
  - Year
  - Gender

- Clean the text data by:
  - Removing punctuation
  - Removing numbers
  - Removing stopwords
  - Applying lemmatization to standardize the language used in reports

### 2. Priority Labeling (Target Variable Creation)

Since the dataset lacks pre assigned priority labels, a **rule based system** using curated keyword lists for High, Medium, and Low priority will be developed. This establishes the "ground truth" for training.

### 3. Feature Engineering

Convert the cleaned text into numerical features using **TF-IDF (Term Frequency-Inverse Document Frequency)** vectorization. This technique captures:
- The importance of words
- Simple phrases (unigrams and bigrams) within the context of the entire dataset

### 4. Model Development and Evaluation

- Train and compare multiple classification algorithms:
  - **Logistic Regression**
  - **Support Vector Machines (SVM)**

These algorithms are well suited for text classification tasks.

- Evaluate model performance using metrics that align with the business goal of capturing urgent issues.

---

## Final Output

The final output is a **predictive system** that can:

1. Take a new student report as input
2. Instantly assign a priority level (High, Medium, or Low)
3. Enable the support team to triage effectively

# Data Preprocessing & Labeling

### Create priority labels by classifying each complaint into one of the following categories:

- HIGH Priority: Critical issues that directly affect academics or urgent well being (e.g., no Wi-Fi, inability to access exams, emergency financial issues, health crises)
- MEDIUM Priority: Important concerns that are not immediately critical (e.g., financial aid inquiries, course availability, difficulty accessing advisors)
- LOW Priority: General feedback, suggestions, or minor inconveniences (e.g., food quality, social activities)

```python
priority_keywords = {
    'HIGH': {
        'technical': ['wifi not working', 'internet down', 'system crash', 'can\'t access', 
                      'login issue', 'technical difficulty', 'server down', 'blackboard down'],
        'academic_emergency': ['failing', 'drop out', 'withdraw', 'grade appeal', 
                               'academic warning', 'probation', 'cannot submit'],
        'financial_emergency': ['homeless', 'eviction', 'no money for food', 
                                'urgent financial', 'emergency loan'],
        'health_emergency': ['medical emergency', 'hospital', 'mental health crisis',
                             'suicidal', 'self-harm', 'crisis']
    },
    'MEDIUM': {
        'financial': ['financial aid', 'scholarship', 'tuition payment', 'loan',
                      'student debt', 'cost of attendance', 'textbook cost'],
        'academic': ['advisor', 'course registration', 'graduation requirement',
                     'degree plan', 'course availability', 'prerequisite'],
        'housing': ['dorm issue', 'housing application', 'roommate conflict',
                    'maintenance request', 'lease issue'],
        'support': ['counseling', 'disability services', 'accommodation',
                    'tutoring', 'academic support']
    },
    'LOW': {
        'general': ['feedback', 'suggestion', 'improvement', 'would like',
                    'wish there were', 'it would be great'],
        'social': ['activities', 'events', 'clubs', 'organizations',
                   'student life', 'social events'],
        'facilities': ['parking', 'library hours', 'gym hours', 'food quality',
                       'cafeteria', 'vending machine']
    }
}


def assign_priority(text):
    text_lower = text.lower()
    
    for category, keywords in priority_keywords['HIGH'].items():
        for keyword in keywords:
            if keyword in text_lower:
                return 'HIGH'
    
    for category, keywords in priority_keywords['MEDIUM'].items():
        for keyword in keywords:
            if keyword in text_lower:
                return 'MEDIUM'
    
    return 'LOW'
```

```python
df['Priority'] = df['Reports'].astype(str).apply(assign_priority)

print("\nPriority Distribution (after labeling):")
print(df['Priority'].value_counts())

sns.countplot(x='Priority', data=df, order=['HIGH', 'MEDIUM', 'LOW'])
plt.title('Distribution of Assigned Priority Levels')
plt.show()
```

- The dataset is **severely imbalanced** (~90% LOW).  
- This makes any ML model biased towards the LOW class.  
- More importantly, the **rule-based labeling is not semantically accurate** — it does not truly understand the meaning of the reports.

### Transformer Based Priority Classification for Reports

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", 
                      model="facebook/bart-large-mnli", 
                      device=-1) 

def get_semantic_priority(report_text):

    if pd.isna(report_text) or report_text.strip() == "":
        return "LOW"

    candidate_labels = [
        "urgent critical emergency help", 
        "important academic or financial inquiry", 
        "general feedback or routine request"
    ]
    
    label_map = {
        "urgent critical emergency help": "HIGH",
        "important academic or financial inquiry": "MEDIUM",
        "general feedback or routine request": "LOW"
    }

    result = classifier(report_text, candidate_labels, multi_label=False)
    
    top_label = result['labels'][0]
    return label_map[top_label]

print("Starting semantic labeling... this may take a few minutes.")
df['Priority'] = df['Reports'].apply(get_semantic_priority)

print("\nNew Semantic Priority Distribution:")
print(df['Priority'].value_counts())

```

```python
plt.figure(figsize=(8, 5))  
sns.boxplot(data=df, x="Priority", y="Gpa", order=["HIGH", "MEDIUM", "LOW"], palette="Set2")
plt.title("GPA Distribution by Semantic Priority")
plt.xlabel("Priority")
plt.ylabel("GPA")
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Priority", order=["HIGH", "MEDIUM", "LOW"], palette="Set2")
plt.title("Final Priority Distribution")
plt.xlabel("Priority")
plt.ylabel("Count")
plt.show()
```

The count of **HIGH** priority reports has tripled (from 6 to 17), and **MEDIUM** has now become the largest category. This indicates that the model is correctly identifying that most student reports contain important academic or financial inquiries, rather than just general feedback.  

However, there is still a significant **class imbalance** (17 HIGH vs. 802 MEDIUM). If a standard model is trained on this dataset  s-is, it will likely struggle to distinguish between **MEDIUM** and **HIGH** priorities, because the overwhelming number of MEDIUM examples will dominate the learning process.

## Advanced Text Preprocessing

```python
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    if not isinstance(text, str):
        return ""
    
    text = text.lower()
    text = re.sub(r'\d+', '', text)                    
    text = text.translate(str.maketrans('', '', string.punctuation))  
    text = re.sub(r'\s+', ' ', text).strip()          
    
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return ' '.join(tokens)

print("Cleaning text...")
df_final['cleaned_reports'] = df_final['Reports'].apply(clean_text)

print("\nSample before/after cleaning:")
for i in range(3):
    print(f"Original: {df_final['Reports'].iloc[i][:150]}...")
    print(f"Cleaned : {df_final['cleaned_reports'].iloc[i][:150]}...\n")
```

## Advanced Transformer Classifier

Currently, with our **hyperparameter-tuned XGBoost**, we have improved the model’s performance.  

To explore whether we can achieve even better results, the next step is to experiment with an **Advanced Transformer-based Classifier**.  

Transformers can:  
- Capture complex patterns in text data  
- Understand semantic relationships better than traditional models  
- Potentially improve classification metrics beyond what XGBoost can achieve  

This approach will help us evaluate if moving to deep learning can provide additional performance gains.

```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.optim import AdamW
from transformers import get_linear_schedule_with_warmup
from sklearn.metrics import classification_report, f1_score
import numpy as np
```

```python
class PriorityDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        return {
            'input_ids':      encoding['input_ids'].squeeze(),
            'attention_mask': encoding['attention_mask'].squeeze(),
            'label':          torch.tensor(self.labels[idx], dtype=torch.long)
        }
```

```python
MODEL_NAME = "distilbert-base-uncased"   
tokenizer  = AutoTokenizer.from_pretrained(MODEL_NAME)

train_texts = df_final.iloc[y_train.index]['Reports'].tolist()
test_texts  = df_final.iloc[y_test.index]['Reports'].tolist()

train_labels_enc = label_encoder.transform(y_train.values)
test_labels_enc  = label_encoder.transform(y_test.values)

train_dataset = PriorityDataset(train_texts, train_labels_enc, tokenizer)
test_dataset  = PriorityDataset(test_texts,  test_labels_enc,  tokenizer)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=16)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

num_labels = len(label_encoder.classes_)   

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels
)
model.to(device)

EPOCHS    = 5
optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=0.01)

total_steps    = len(train_loader) * EPOCHS
warmup_steps   = int(0.1 * total_steps)   

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=warmup_steps,
    num_training_steps=total_steps
)
```

```python
def train_epoch(model, loader, optimizer, scheduler, device):
    model.train()
    total_loss, correct, total = 0, 0, 0

    for batch in loader:
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels         = batch['label'].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)

        loss = outputs.loss
        loss.backward()

        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # prevent exploding gradients

        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        preds        = outputs.logits.argmax(dim=1)
        correct     += (preds == labels).sum().item()
        total       += labels.size(0)

    return total_loss / len(loader), correct / total


def eval_epoch(model, loader, device):
    model.eval()
    all_preds, all_labels = [], []

    with torch.no_grad():
        for batch in loader:
            input_ids      = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels         = batch['label'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            preds   = outputs.logits.argmax(dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    f1 = f1_score(all_labels, all_preds, average='weighted')
    return f1, all_preds, all_labels

```

```python
print("Training DistilBERT...\n")
for epoch in range(EPOCHS):
    train_loss, train_acc = train_epoch(model, train_loader, optimizer, scheduler, device)
    val_f1, _, _          = eval_epoch(model, test_loader, device)

    print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | Val F1: {val_f1:.4f}")
```

```python
final_f1, final_preds, final_labels = eval_epoch(model, test_loader, device)

print("\nDistilBERT Classification Report:")
print(classification_report(
    final_labels,
    final_preds,
    target_names=label_encoder.classes_
))

comparison = {
    'Model':    ['XGBoost (Original)', 'XGBoost (Tuned)', 'DistilBERT'],
    'F1-Score': [0.9225, 0.9351, round(final_f1, 4)]
}

df_compare = pd.DataFrame(comparison)
colors = ['skyblue', 'orange', 'mediumseagreen']

plt.figure(figsize=(8, 5))
bars = plt.bar(df_compare['Model'], df_compare['F1-Score'], color=colors)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.002,
             f"{bar.get_height():.4f}", ha='center', fontweight='bold')

plt.title('Model Comparison: XGBoost vs DistilBERT')
plt.ylabel('Weighted F1-Score')
plt.ylim(0.88, 1.0)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
